# Export to Clips deffacts
This notebook shall hold the code to read in the webscraped csv and export into the deffacts clips format. <br>
It will be exported in a .txt file so that the user can easily copy and paste it in the .clp file

```python
import pandas as pd
```

```python
df = pd.read_csv('FinalGivingSGScrape20Feb.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
f = open("Output.txt", "w")
```

```python
for i in range(len(df[:1])):
    print('''(event (name-org "{}-{}") (start-date "{}") (end-date "{}") (duration {}) 
         (wkend-wkday-any "{}") (location "{}") (min-age {}) (opening {})
         (cause "{}" "{}" "{}") (skill "{}" "{}" "{}") (suitability "{}" "{}" "{}" "{}"))'''.format(df['Event Name'][i], 
                                                                                 df['Organisation'][i], 
                                                                                 df['Start Date'][i],
                                                                                df['Last Date'][i],
                                                                                df['Duration (Hours)'][i],
                                                                                df['Weekend/Weekday/Anyday'][i],
                                                                                df['Location'][i],
                                                                                 df['Min Age'][i],
                                                                                 df['Openings'][i],
                                                                                df['Cause1'][i],
                                                                                df['Cause2'][i],
                                                                                df['Cause3'][i],
                                                                                df['Skills1'][i],
                                                                                df['Skills2'][i],
                                                                                df['Skills3'][i],
                                                                                df['Suitability1'][i],
                                                                                df['Suitability2'][i],
                                                                                df['Suitability3'][i],
                                                                                df['Suitability4'][i]))
```

```python
for i in range(len(df)):
    f.write('''(event (name-org "{}-{}") (start-date "{}") (end-date "{}") (duration {}) 
         (wkend-wkday-any "{}") (location "{}") (min-age {}) (opening {})
         (cause "{}" "{}" "{}") (skill "{}" "{}" "{}") (suitability "{}" "{}" "{}" "{}"))\n'''.format(df['Event Name'][i], 
                                                                                 df['Organisation'][i], 
                                                                                 df['Start Date'][i],
                                                                                df['Last Date'][i],
                                                                                df['Duration (Hours)'][i],
                                                                                df['Weekend/Weekday/Anyday'][i],
                                                                                df['Location'][i],
                                                                                 df['Min Age'][i],
                                                                                 df['Openings'][i],
                                                                                df['Cause1'][i],
                                                                                df['Cause2'][i],
                                                                                df['Cause3'][i],
                                                                                df['Skills1'][i],
                                                                                df['Skills2'][i],
                                                                                df['Skills3'][i],
                                                                                df['Suitability1'][i],
                                                                                df['Suitability2'][i],
                                                                                df['Suitability3'][i],
                                                                                df['Suitability4'][i]))
f.close()
```
